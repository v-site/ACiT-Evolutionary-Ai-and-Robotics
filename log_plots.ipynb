```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
filenames = ['42_15_7_5_200_500_100_3_6_uniform_equal_split_0.01_0.0_0.9_03.11.2022_11.51.31.txt',
             '42_15_7_5_200_500_100_3_6_uniform_equal_split_0.01_0.1_0.8_02.11.2022_14.49.19.txt',
             '42_15_7_5_200_500_100_3_6_uniform_equal_split_0.01_0.2_0.7_02.11.2022_14.48.17.txt',
             '42_15_7_5_200_500_100_3_6_uniform_equal_split_0.01_0.3_0.6_02.11.2022_14.38.05.txt',
             '42_15_7_5_200_500_100_3_6_uniform_equal_split_0.01_0.4_0.5_02.11.2022_14.38.29.txt']
df_names =['00e90m','10e80m','20e70m','30e60m','40e50m']  
log_dir_path = os.path.join('logs')
dfs =[]
i =0
for fileName in filenames:
    
    load_path = os.path.join(log_dir_path,fileName)
    tmp_df = pd.read_csv(filepath_or_buffer=load_path)
    tmp_df.name =df_names[i]
    dfs.append(tmp_df)
    i+=1



```

```python
plt.rcParams['figure.figsize'] = 40,10
for df in dfs:
    plt.plot(df['maxR'],label = df.name + ' max reward', linestyle = '-' )
    plt.plot(df['avgR'],label = df.name + ' avg reward', linestyle = '--')
plt.legend()
plt.show()

```
